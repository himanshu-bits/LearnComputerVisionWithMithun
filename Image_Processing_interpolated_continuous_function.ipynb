```python
from scipy.interpolate import RectBivariateSpline
import numpy as np

# Small intensity matrix
I = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

# Define grid for discrete data
x = np.arange(1, I.shape[1] + 1)
y = np.arange(1, I.shape[0] + 1)

# Reduce the degree of the spline to 1 (linear interpolation)
interp_func = RectBivariateSpline(y, x, I, kx=1, ky=1)

# Evaluate the function at continuous coordinates
x_cont = np.linspace(1, 3, 100)
y_cont = np.linspace(1, 3, 100)
I_cont = interp_func(y_cont, x_cont)
```

```python
# Plot the continuous function
plt.imshow(I_cont, extent=(1, 3, 1, 3), origin='lower', cmap='gray')
plt.colorbar(label='Intensity')
plt.title("Continuous Intensity Function")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import numpy as np
from scipy.interpolate import RectBivariateSpline
import matplotlib.pyplot as plt

# Small intensity matrix (original image)
I = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

# Define grid for discrete data
x = np.arange(1, I.shape[1] + 1)
y = np.arange(1, I.shape[0] + 1)

# Reduce the degree of the spline to 1 (linear interpolation)
interp_func = RectBivariateSpline(y, x, I, kx=1, ky=1)

# Evaluate the function at continuous coordinates
x_cont = np.linspace(1, 3, 100)
y_cont = np.linspace(1, 3, 100)
X, Y = np.meshgrid(x_cont, y_cont)
I_cont = interp_func(y_cont, x_cont)

# Plot original matrix and interpolated function
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Original image
axes[0].imshow(I, extent=(1, 3, 1, 3), origin='lower', cmap='gray', interpolation='none')
axes[0].set_title("Original Intensity Matrix")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")

# Interpolated continuous function
axes[1].imshow(I_cont, extent=(1, 3, 1, 3), origin='lower', cmap='gray')
axes[1].set_title("Interpolated Continuous Function")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")

plt.tight_layout()
plt.show()
```
